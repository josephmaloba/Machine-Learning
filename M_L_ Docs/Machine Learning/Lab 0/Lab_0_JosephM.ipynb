<h2>Question 1</h2>

<h3>Write a Python script that asks the user to enter a year between 1 and 3199, inclusive, and  
tells the user whether that was or will be a leap year or not</h3>


```python
def leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

def get_valid_year():
    while True:
        try:
            year = int(input("Enter a year between 1 and 3199: "))
            if 1 <= year <= 3199:
                return year
            else:
                print("Year must be between 1 and 3199. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

# Main program
year = get_valid_year()

if leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

<h3> Question 2</h3>

<h3>Use a while loop to present the user with a multiple-correct multiple-choice question and 
evaluate their response</h3>>

```python
def multiple_choice_quiz():
    question = "Which of the following countries can be considered Scandinavian?"
    options = {
        'A': 'Latvia',
        'B': 'Denmark',
        'C': 'Finland',
        'D': 'The Netherlands',
        'E': 'Sweden'
    }
    correct_answers = ['B','C','D']

    print(question)
    for key, value in options.items():
        print(f"{key}. {value}")

    user_answers = []
    while True:
        answer = input("Enter your answer (one letter at a time) or 'done' when finished: ").upper()
        if answer == 'DONE':
            break
        if answer in options:
            user_answers.append(answer)
        else:
            print("Invalid input. Please enter a valid option (A, B, C, D, or E) or 'done'.")

    correct_count = sum(1 for ans in user_answers if ans in correct_answers)
    incorrect_count = len(user_answers) - correct_count

    print("\nResults:")
    print(f"Correct answers: {correct_count}")
    print(f"Incorrect answers: {incorrect_count}")
    
    if set(user_answers) == set(correct_answers):
        print("You got all the correct answers.")
    elif correct_count > 0:
        print("You got some correct answers, but not all of them.")
    else:
        print("Sorry, you didn't get any correct answers.")

    print(f"\nThe correct answers were: {', '.join(correct_answers)}")
    
    return correct_count, incorrect_count

# Run the quiz and get the results
correct, incorrect = multiple_choice_quiz()
print(f"\nFinal score - Correct: {correct}, Incorrect: {incorrect}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('https://archive.ics.uci.edu/static/public/409/daily+demand+forecasting+orders.zip', sep=";")

print(df.columns)

# Calculate range, average, and variance for the 12 features
for column in df.columns[:-1]:  # Exclude the last column (target)
    print(f"\n{column}:")
    print(f"Range: {df[column].max() - df[column].min()}")
    print(f"Average: {df[column].mean()}")
    print(f"Variance: {df[column].var()}")

# Calculate range, average, and variance for the target
target = df.columns[-1]
print(f"\n{target} (Target):")
print(f"Range: {df[target].max() - df[target].min()}")
print(f"Average: {df[target].mean()}")
print(f"Variance: {df[target].var()}")

# Create a scatter plot of 'Week of the month' against 'Target'
plt.figure(figsize=(10, 6))
plt.scatter(df['Week of the month (first week, second, third, fourth or fifth week'], df[target])
plt.xlabel('Week of the month')
plt.ylabel(target)
plt.title('Scatter plot: Week of the month vs Target')
plt.show()

# Calculate correlation coefficient
corr = np.corrcoef(df['Week of the month (first week, second, third, fourth or fifth week'], df[target])[0, 1]
print(f"\nCorrelation coefficient between 'Week of the month' and '{target}': {corr}")
```
